```python
def count_lines(filename):
    try:
        with open(filename, 'r') as file:
            line_count = 0
            for line in file:
                line_count += 1
        return line_count
    except FileNotFoundError:
        print("File not found.")
        return -1

filename = "ICT110_PythonWorkbook"
num_lines = count_lines(filename)
if num_lines != -1:
    print(f"Number of lines in {filename}: {num_lines}")
```

```python
def print_file_uppercase(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                print(line.upper(), end='')
    except FileNotFoundError:
        print("File not found.")

filename = "ICT110_PythonWorkbook"
print_file_uppercase(filename)
```

```python
def extract_spam_confidence(filename):
    try:
        with open(filename, 'r') as file:
            total_confidence = 0.0
            count = 0
            for line in file:
                if line.startswith("X-DSPAM-Confidence:"):
                    try:
                        confidence = float(line.split(':')[1].strip())
                        total_confidence += confidence
                        count += 1
                    except ValueError:
                        print("Error converting confidence to float.")
            if count > 0:
                average_confidence = total_confidence / count
                print(f"Average spam confidence: {average_confidence}")
            else:
                print("No spam confidence values found in the file.")
    except FileNotFoundError:
        print("File not found.")

filename = input("Enter the file name: ")
extract_spam_confidence(filename)
```
