```python
filename = input('Enter a filename:')
try:
  with open(filename, 'r') as file:
    for line in file:

        line = line.strip()

        if line.startswith('From '):
            words = line.split()
            print(words[2])

except FileNotFoundError:
        print("File not found. Please make sure the file exists and try again.")
```
