**Exercise 2**

```python
gdp_growth = {
    'China': 6.5,
    'Japan': 0.7,
    'Philippines': 6.7,
    'United States': 2.3
}

print("Keys:")
for key in gdp_growth:
    print(key)

print("\nValues:")
for value in gdp_growth.values():
    print(value)

print("\nKeys and Values:")
for key, value in gdp_growth.items():
    print(f"{key}: {value}")
```

**Exercise 3**

```python
def count_words(line):
    word_list = line.split()
    word_counts = {}

    for word in word_list:
        word_counts[word] = word_counts.get(word, 0) + 1

    return word_counts

line = input("Enter a line of text: ")

word_counts = count_words(line)

print("Word Counts:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```
